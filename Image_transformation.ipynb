```python
import torch
import torch.nn as nn
import torch.optim as optim

# define a single model
class SingleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)

    def forward(self, x):
        return self.linear(x)

# Data
X = torch.tensor([[1.0,2.0],[2.0,3.0]])
y = torch.tensor([[1.0],[2.0]])

# model, loss, optimizer
model = SingleModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01) #lr - lerning rate

# training loop
for epoch in range(100):
    # forward pass
    y_pred = model(X)

    # compute loss
    loss = criterion(y_pred, y)

    # backward pass
    loss.backward()

    # update weights
    optimizer.step()

print(f'Trained weights:{list(model.parameters())}')
```
